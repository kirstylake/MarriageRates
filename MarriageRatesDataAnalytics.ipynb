# 3 Specific data and correlation analysis

 ## 3.1 Check the entire dataset to see if we can make the assumption that in general the crude marriage rate is decreasing over time

In the analysis below, I will be looking for a way to generalize whether each entity is either increasing their crude marriage rate (per 1000 inhabitants) over time, or if they are decreasing over time. I have decided to do this by calculating the slope between the earliest record for an entity and the latest record of an entity. I am using this to create a trend line and from there calculate whether the slope is increasing or decreasing. I then need to do a deeper analysis on the spread of this data accross the world to see if it matches any other common data like wealth spread and educational spread. If there is no correlation there, then I will need to conduct more research on what could be causing the data to be spread the way that it is.

#### Calculate Slope

```python
# There will be warnings due to divides by zero but these are not necessary for display as their result is removed
# from the dataset
import warnings
warnings.filterwarnings('ignore')

# The slope of all the entities need to be calculated to determine 
# if we can make the assumption that in general marraiage rates are going down accross the world 

# This is a function to determine the gradient of the slope
def slopeCalc(x1,x2,y1,y2):
    x = (y2 - y1)/ (x2 - x1)
    return x

# All the unique entities are being found here to be examined
Entities = df['Entity'].unique()
slopeTable = pd.DataFrame(columns=['Entity','Slope'])

# The unique entities are being looped through to find the slope points, we are looking
# at the slope from the earliest time point to the latest time point and finding the associated CMR values for each of these

for entity in Entities:
#   Here we create a new dataframe just for each unique entity
    dfTemp = df[df['Entity'] == entity]

#   Then we are finding the minimum year and the maximum year
    maxYear = dfTemp['Year'].max()
    minYear = dfTemp['Year'].min()
    
#   Now we find the assosciated CMR for each of these years
    dfTemp2 = dfTemp[dfTemp['Year'] == minYear]
    CMR1 = dfTemp2['CMR'].values[0]
    
    dfTemp2 = dfTemp[dfTemp['Year'] == maxYear]
    CMR2 = dfTemp2['CMR'].values[0]
    
#     print('Entity ' + str(entity) + ' MaxYear ' + str(maxYear) + ' MinYear ' + str(minYear) + ' C1 ' + str(CMR1) + ' C2 ' + str(CMR2) )
    
#   We calculate the slope now for each of these values
    slope = slopeCalc(CMR1,CMR2,minYear,maxYear)
    
#   Add the data to another data frame
    new_row = pd.DataFrame({'Entity':[entity], 'Slope': [slope]})
    slopeTable = pd.concat([slopeTable, new_row], ignore_index = True, axis = 0 )                                                

    
# some of the data elements had only one entry so they returned a NaN value, 
# we need to cleanse our data of these values
slopeTable = slopeTable.dropna()

# Now we have a database of slopes associated with Entities
slopeTable   
```

Now we need to find out how many of the records have a negative slope so we can see if there is a correlation present

```python
len(slopeTable)
```

#### Calculate distribution of slope

There are 169 Entities, each with a valid slope calculation, now we need to find out how many of these are negative and how many of these are positive.

```python
sumNeg = 0;
sumPos = 0;

# create a list object to store the names of the countries that have increased in marriage rates over time
PosList = [];

# create a list object to store decreasing values
NegList = [];

# iterating over the dataframe to find how many marriage rates have increased or decreated over time
for index, row in slopeTable.iterrows():
    if row['Slope'] < 0:
        sumNeg = sumNeg + 1
    else:
        sumPos = sumPos + 1
        PosList.append(row['Entity'])
        
print('The total amount of countries where marriage rates have declined is ' + str(sumNeg))
print('The total amount of countries where marriage rates have increased is ' + str(sumPos))

# print(PosList)
```

This is an interesting result, I originally thought that the amount of countries decreasing in marriages over time would be higher. The dataset does however include non-western societies which follow different cultural norms that originally wasn't in the scope of the report.

#### Visualize Spread with a Pie Chart

The result can be visualized with a pie chart

```python
# green for increasing and red for decreasing
colors = ( "green", "red")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

percentages = [sumPos/len(slopeTable) * 100, sumNeg/len(slopeTable) * 100]

labels = ['Increasing CMR', 'Decreasing CMR']

# # Creating autocpt arguments
# This creates the labels on the wedges of the graph
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%".format(pct, absolute)

# The settings for how the pie chart should be displayed
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(percentages,
                                  autopct = lambda pct: func(pct, percentages),
                                  labels = labels,
                                  shadow = False,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black"))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Pie Chart showing the proportion of entities with a declining or increasing Crude Marriage Rate (CMR) per 1000 inhabitants")

plt.show()
```

#### Visualize Spread of marriage data on a map

```python
# Install geopandas using the command conda install -c conda-forge geopandas in the Anaconda prompt
# This code was partially adapted from [2]
# You may need to install some dependencies if you are on windows, you need to acces the library database [4]
# You will need to install GDAL, Pyproj, Fiona, Shapely and Geopandas as shown in [5]

import geopandas       

# get the path of the geopandas standard map
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

# Increase the size of the map
f, ax = plt.subplots(1, figsize=(12, 12))
ax.set(title='Visualization of the spread of changing marriage rates in the world')
ax = world.plot(axes=ax)


# some of the values for this will not show on the map just because the country name may not
# match the country name specified by geopandas
for index, row in slopeTable.iterrows():
    if row['Slope'] > 0:
        try:
            world[world.name == row['Entity']].plot(color='green',ax=ax)
        except:
            print(row['Entity'] + ' is not in the geopandas country names')
    if row['Slope'] < 0:
        try:
             world[world.name == row['Entity']].plot(color='red',ax=ax)
        except:
            print(row['Entity'] + ' is not in the geopandas country names')
plt.show()

```
